```python
import pickle
import pathlib
import numpy as np
```

```python
DATA_DIR = pathlib.Path.cwd().parent / 'data'
RAW_DATA_DIR = DATA_DIR / 'raw'
PDF_DIR = RAW_DATA_DIR / 'pdf'
DF_DIR = RAW_DATA_DIR / 'dataframe'
```

```python
with open(DF_DIR / 'data.pkl', 'rb') as f:
    raw_data = pickle.load(f)
```

```python
raw_data.keys()
```

```python
data = raw_data.copy()
example = data['2021-12']
example
```

```python
# Delete unused columns
unused_columns = ['company', 'industry', 'upside', 'market_cap']
for key in data.keys():
    data[key].drop(columns=unused_columns, inplace=True)
```

```python
example
```

```python
for df in data.values():
    for column in ['price', 'target']:
        df[column] = df[column].apply(lambda x: x.replace('R$', '').replace(',', '.'))
        df[column] = df[column].apply(lambda x: np.nan if x == '-' or x == 'Restrito' else x)
        df[column] = df[column].astype(float)
```

```python
for df in data.values():
    df['portfolio_weight'] = df['portfolio_weight'].apply(lambda x: x.replace('%', '').replace(',', '.'))
    df['portfolio_weight'] = df['portfolio_weight'].astype(float)
    df['portfolio_weight'] = df['portfolio_weight'] / 100
```

```python
for df in data.values():
    df['ticker'] = df['ticker'].apply(lambda x: x + '.SA')
    df['ticker'] = df['ticker'].astype(str)
```
